```python
system_prompt = """You are an AI Time Table Generator for 3rd year B.Tech students.
Your job is to create a structured daily timetable for students who are managing:
-College subjects
-Labs
-Workshops
-Placement preparation
-Coding practice
-Revision
-Projects

Rules:
1. The timetable should be balanced.
2. Include breaks.
3. Include at least: 2 hours subject study, 1 hour coding/DSA, 1 hour revision
4. Sleep should be 7 hours minimum.
5. Make the timetable in table format.
6. The timetable should be realistic and consistent."""
```

```python
college_time = input("Enter college timing (e.g., 9am-2pm): ")
subjects = input("Enter your subjects: ")
workshop = input("Workshop time (if any): ")
sleep_time = input("When do you sleep? ")
weak_subject = input("Your weak subject: ")

print("\nCollected Information:")
print("College Time:", college_time)
print("Subjects:", subjects)
print("Workshop:", workshop)
print("Sleep Time:", sleep_time)
print("Weak Subject:", weak_subject)
```

```python
user_prompt = f"""
Create a daily timetable for a 3rd year B.Tech student.

Details:
College Time: {college_time}
Subjects: {subjects}
Workshop: {workshop}
Sleep Time: {sleep_time}
Weak Subject: {weak_subject}

Generate a structured timetable in table format.
"""
```

```python
user_prompt
```

```python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

```python
chat_response = openai.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=messages,
    temperature=0.3,
    max_tokens=500
)
```
