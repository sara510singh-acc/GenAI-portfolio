```python
import openai
```

```python
openai.api_key="---------------------------------------------------------------------------------------------------------"
```

```python
messages = [
        {"role": "system", "content": "You are a Data Science assistant.You ONLY answer questions related to Data Science. If a question is NOT related to Data Science, respond ONLY with: 'Sorry! I can't answer this question. I can answer only Data Science related stuff.'"},
        {"role": "user", "content": "What is data science?"}
    ]
```

```python
messages
```

```python
chat_response = openai.chat.completions.create(
        model="gpt-3.5-turbo-16k",
        messages=messages
    )
```

```python
chat_response
```

```python
chat_response.choices[0].message.content
```
