```python
from transformers import pipeline #downloads all the hugging face models, pipeline is a parameter.
```

```python
classifier = pipeline ("sentiment-analysis") #picking a sentiment analysis model
```

```python
classifier("the food was delicious")
```

```python
classifier("I wish food was good.")
```

```python
classifier("I am amazed by the speed of the processor but disappointed that it heats up quickly.")
```

```python
classifier(["I've been waiting for a HuggingFace course my whole life.", "I hate this so much.","She was disappointed but she scored well."])
```
